# Study Crash Predictor - Data Exploration

This notebook explores the synthetic dataset, visualizes distributions,
and analyzes relationships between features and the target variable.

```python
import sys, os
sys.path.insert(0, os.path.dirname(os.path.dirname(os.path.abspath('__file__'))))

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

from src.generate_data import generate_dataset
```

```python
# Generate dataset
df = generate_dataset(n_samples=5000, seed=42)
print(f'Shape: {df.shape}')
df.head()
```

```python
# Basic statistics
df.describe()
```

```python
# Class distribution
print(df['target'].value_counts())
df['target'].value_counts().plot(kind='bar', color=['#38ef7d', '#f5576c', '#ff4b2b'])
plt.title('Target Class Distribution')
plt.ylabel('Count')
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

```python
# Feature distributions by target class
features = [c for c in df.columns if c != 'target']

fig, axes = plt.subplots(3, 4, figsize=(20, 12))
axes = axes.flatten()

for i, col in enumerate(features):
    for target_val in df['target'].unique():
        subset = df[df['target'] == target_val]
        axes[i].hist(subset[col], alpha=0.5, label=target_val, bins=20)
    axes[i].set_title(col, fontsize=10)
    axes[i].legend(fontsize=7)

for j in range(len(features), len(axes)):
    axes[j].set_visible(False)

plt.suptitle('Feature Distributions by Target Class', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap
from sklearn.preprocessing import LabelEncoder

df_encoded = df.copy()
df_encoded['target'] = LabelEncoder().fit_transform(df_encoded['target'])

plt.figure(figsize=(12, 8))
sns.heatmap(df_encoded.corr(), annot=True, fmt='.2f', cmap='RdBu_r', center=0)
plt.title('Feature Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
# Pairplot for key features
key_features = ['hours_slept', 'phone_pickups_last_hour', 'mood_level', 'study_duration_minutes', 'target']
sns.pairplot(df[key_features], hue='target', diag_kind='kde', plot_kws={'alpha': 0.4, 's': 15})
plt.suptitle('Pairplot of Key Features', y=1.02)
plt.show()
```
